```python
import pandas as pd
import numpy as np
import seaborn as sns

df = sns.load_dataset('titanic')
```

.() function used to read the top few rows in a data group

```python
df.head() 
```

describe() function used to get some general info on the data set

```python
df.describe()
```

loop to print the .unique() results in each column, to see the variety of values, and especially null values

```python
for col in df:
  print(df[col].unique())
```

dropna() drops all the Null values

```python
new_df = df.dropna()
new_df.head()
```

However, we can see that the count has dropped from 891 to 182, so is too severe...

```python
new_df.count()
```

duplicated() method checks for duplicated rows

```python
df[df. duplicated()]
```

then, drop_duplicates() is used to take these out
(inplace=True) here specifies that this change should be made permanently to the original DataFrame, rather than in a copy

```python
df.drop_duplicates(inplace = True)
```

groupby() method is used to group by a column, as in SQL
mean() to generate an average

```python
df.groupby('class')['fare', 'age'].mean().round()
```

len() returns values based on column labels (like SELECT in SQL)... here we can return all the rows where sex = female

```python
len(df[df['sex']=='female'])
```

adding this method into a new dictionary, we can then create a new DataFrame with the data

```python
women_and_men = {
    "total women": [len(df[df['sex']=='female'])],
    "total men": [len(df[df['sex']=='male'])]
}
df2 = pd.DataFrame(women_and_men)
df2
```

we can return the total fare using the .sum() method
(skipna=True) here specifies that all Null values should be ignored

```python
total_fare = df['fare'].sum(skipna=True)
total_fare
```

we can then use the .apply() method to add in another column, in which we caluclate the percentage of total fare using the total_fare variable from above

```python
df['Percentage of Total Fare'] = df.apply(lambda row: (row['fare']/total_fare * 100).round(4), axis=1)
df.head()
```
